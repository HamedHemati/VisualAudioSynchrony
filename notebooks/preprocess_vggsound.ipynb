```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import random
```

```python
def get_frame(video_cap, sec):
    fps = video_cap.get(cv2.CAP_PROP_FPS)
    video_cap.set(1, int(sec * fps))
    success, image_frame = video_cap.read()
    if success:
        return image_frame
    return None

def get_frame_random(video_cap, start, offset):
    fps = video_cap.get(cv2.CAP_PROP_FPS)
    low = int(start * fps)
    high = int((start + offset) * fps)
    max_frames = video_cap.get(cv2.CAP_PROP_FRAME_COUNT)
    if high >= max_frames:
        high = max_frames - 1
    rand_fr = random.randint(low, high)
    
    video_cap.set(1, rand_fr)
    success, image_frame = video_cap.read()
    if success:
        return image_frame
    return None
```

```python
def show_frame(cv2_img):
    img = cv2.cvtColor(cv2_img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)

def save_frame(cv2_img, out_path):
    cv2.imwrite(out_path, cv2_img)
```

```python
video_path = "/netscratch/hhemati/Datasets/MultiModal/VGGSound/videos/video_5780.mp4"
video_cap = cv2.VideoCapture(video_path)
print(video_cap.get(cv2.CAP_PROP_FRAME_COUNT))
```

```python
frame = get_frame_random(video_cap, start=30, offset=10)
if frame is not None:
    show_frame(frame)
    save_frame(frame, "image.jpg")
```
